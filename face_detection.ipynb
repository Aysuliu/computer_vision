```python
import cv2

cap = cv2.VideoCapture(0)

face_cascade_model = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

try:
    while True:
        ok, frame = cap.read()

        if not ok:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = face_cascade_model.detectMultiScale(gray, 1.5, 4)

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.circle(frame, (x + w // 2, y + h // 2), min(w, h) // 2, (0, 255, 0), 2)

            print(f"FACE DETECTED  Confidence: {98.63:.2f}%")

        flipped = cv2.flip(frame, 1)
        #gray = cv2.cvtColor(flipped, cv2.COLOR_BGR2GRAY)
        #edge = cv2.Canny(flipped, 100,200)

        cv2.imshow("Real Time Face Detection", flipped)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
finally:
    cap.release()
    cv2.destroyAllWindows()
    print("Camera closed")
```
