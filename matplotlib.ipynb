```python
import matplotlib.pyplot as plt
import numpy as np
```

# Lazy Line Plot

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)
plt.title("Bar Chart Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Line Styles and Colors

```python
x = np.array([1,2,3,4])
y = np.array([3,8,1,10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Bar Chart Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Multiple Lines in One Graph

```python
x = np.array([1,2,3,4])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(x,y1, label='line 1', color='green', marker='s')
plt.plot(x,y2, label='line 2', color='red', linestyle='--', marker='o')
plt.title("Multiple lines Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Bar Chart with Numpy & Matplotlib

```python
x = np.array(["Salary","Expense","Charity","Food"])
y = np.array([3,8,1,10])

plt.bar(x,y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Scatter Plot with Numpy & Matplotlib

```python
x = np.array([5,7,8,7])
y = np.array([99,86,87,88])

plt.scatter(x,y, color="purple", s=400)
plt.title("Scatter Plot Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Histogram with Numpy & Matplotlib

```python
data = np.array([3, 4, 5, 6, 8,8, 8, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("values")
plt.ylabel("count")
plt.show()
```

# Subplots (Multiple Graphs Together)

```python
x = np.array([1,2,3,4])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(x,y1, color='red', marker='s')
plt.title("Subplot 1")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.subplot(1,2,2)
plt.bar(x,y2, color='blue', linestyle='--')
plt.title("Subplot 2")
plt.tight_layout()
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
plt.style.use('_mpl-gallery')

X, Y, Z = axes3d.get_test_data(0.05)

fig , ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.plot_surface(X, Y, Z, rstride=10, cstride=10)


ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
plt.style.use('_mpl-gallery')
#make data
np.random.seed(19680001)
n = 100
rng = np.random.default_rng()
xs= rng.uniform(23,32,n)
ys= rng.uniform(10,100,n)
zs= rng.uniform(-50,-25,size=n)

#plot
fig , ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.scatter(xs, ys, zs)


ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
