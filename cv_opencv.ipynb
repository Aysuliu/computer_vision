```python
import cv2
```

```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat.jpeg")
cv2_imshow(image)
```

```python
print("Shape:", image.shape)
# 182=>height
# 277=>width
# 3=>color channels(BGR)=>blue,green,red

print("height:", image.shape[0])
print("width:", image.shape[1])
print("channels:", image.shape[2])
```

```python
(b,g,r) = image[100,50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #HLS,RGB
cv2_imshow(gray)
```

```python
resized = cv2.resize(grey, (500,300)) #width, height
cv2_imshow(resized)
```

```python
cropped = grey[100:300, 150:350]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(image, 0)
cv2_imshow(flipped)
```
